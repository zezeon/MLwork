```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('lab2_test.csv')
print(data)
```

```python
class_0 = data[data['class'] == 'Iris-setosa']
class_1 = data[data['class'] == 'Iris-versicolor']
class_2 = data[data['class'] == 'Iris-virginica']

mean_vector_0 = class_0.iloc[:, :-1].mean().values
mean_vector_1 = class_1.iloc[:, :-1].mean().values
mean_vector_2 = class_2.iloc[:, :-1].mean().values

print("Class Iris-setosa Mean Vector:", mean_vector_0)
print("Class Iris-versicolor Mean Vector:", mean_vector_1)
print("Class Iris-virginica Mean Vector:", mean_vector_2)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(class_0['petal_width'], class_0['sepal_length'], label='Iris-setosa', color='r')
plt.scatter(class_1['petal_width'], class_1['sepal_length'], label='Iris-versicolor', color='g')
plt.scatter(class_2['petal_width'], class_2['sepal_length'], label='Iris-virginica', color='b')

plt.title('Scatter Plot of Iris Classes')
plt.xlabel('Petal width')
plt.ylabel('Sepal length')

plt.legend()

plt.show()
```
